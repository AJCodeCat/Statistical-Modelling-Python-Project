Join the data from Part 1 with the data from Part 2 to create a new dataframe.

```python
import pandas as pd
# Load dfs from yelp_fsq notebook

yelp_df = pd.read_csv('C:\Career\LighthouseLabs\PythonProject_Week5\yelp_df.csv', index_col=0)
fsq_df = pd.read_csv(r'C:\Career\LighthouseLabs\PythonProject_Week5\fsq_df.csv', index_col=0)


# Merge yelp_df and fsq_df
#print (POI_df = pd.merge(yelp_df, fsq_df, left_on='latitude', right_on='latitude1', how='left').drop('latitude1', axis=1))

POI_df = pd.merge(yelp_df, fsq_df, left_on='latitude', right_on='latitude', how='left')
print(POI_df)
```

```python
POI_df.to_csv('C:\Career\LighthouseLabs\PythonProject_Week5\POI_df.csv')
```

```python
# POI Duplicates
# Find duplicate rows
duplicate_mask = POI_df.duplicated(keep=False)

# View duplicate rows
duplicate_rows = POI_df[duplicate_mask]
#print(duplicate_rows)

# Delete duplicate rows
POI_df.drop_duplicates(inplace=True)
```
