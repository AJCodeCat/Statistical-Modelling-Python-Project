```python
# Create a Simple Lin Reg using Statsmodel.
import statsmodels.api as sm
```

```python
# Add a Beta-naught y-intercept constant
X = sm.add_constant(X) # adding a constant
lin_reg = sm.OLS(y,X)
```

```python
# Model summary
model = lin_reg.fit()
print_model = model.summary()
print(print_model)
```

```python
import matplotlib.pyplot as plt

# THREE-WAY SCATTERPLOT containing the variables total_bikes, rating, and review_count
plt.scatter(bikePOI1_df['total_bikes'], bikePOI1_df['rating'], c=bikePOI1_df['review_count'], cmap='viridis')
plt.xlabel('Total Bikes')
plt.ylabel('Rating')
plt.colorbar(label='Review Count')
plt.title('Three-Way Scatter Plot')
plt.show()
```
